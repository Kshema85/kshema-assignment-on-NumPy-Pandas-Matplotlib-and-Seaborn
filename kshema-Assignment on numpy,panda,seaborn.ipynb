# To plot the classification of species  data using a box plot

```python
sns.boxplot(x="Classification", y="SL", data=iris )
plt.show()
```

# plot histogram -SL 

```python
fig, ax = plt.subplots()

ax.hist(iris['SL'])
# set title and labels
ax.set_title('Sepallength')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

# Creating the bar chart 

```python
iris1 = iris.groupby('Classification')['SW'].sum().to_frame().reset_index()

plt.bar(iris1['Classification'],iris1['SW'],color = ['blue','lightseagreen','steelblue']) 

plt.title('Bar Chart OF IRIS ')
plt.xlabel('Classification')
plt.ylabel('sepalwidth') 

plt.show()
```

# Pair plot of iris dataset

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris,hue="Classification",height=, diag_kind="kde")
plt.show()
```
